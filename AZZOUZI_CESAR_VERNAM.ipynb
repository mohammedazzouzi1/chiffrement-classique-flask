## 1. Bases — Encodage ASCII
> Avant de chiffrer, on doit comprendre comment les lettres sont représentées numériquement.
> L'alphabet majuscule commence à ASCII 65 (A=65, B=66, ... Z=90).
> On **normalise sur 26** en soustrayant 65 : A=0, B=1, ... Z=25.

### 1.1 Affichage de la valeur ASCII d'un caractère

```python
print("--------affichage ASCII---------")
var=input("entrer le caractere en majuscule: ")
print("la valeur ASCII de ",ord(var))
print("voici son code ASCCI normalise sur 26:",ord(var)-65)
```

### 1.2 Chiffrement d'un seul caractère avec une clé fixe (clef = 2)
> On décale le code ASCII normalisé de la valeur de la clé.

```python
print("----------ajout clef=2-----------")
clef=2
var=input("entrer le caractere en majuscule: ")
ciphred=ord(var)-65+int(clef)
print("le caractere chiffré est: ",ciphred)
print("voici la lettre chiffree :",chr(ciphred+65))
```

### 1.3 Gestion du dépassement de Z — Opération modulo 26
> Problème : si Z (25) + clef = 27, on sort de l'alphabet.
> Solution : appliquer **% 26** pour revenir à A (0).

```python
print('----------test du cas de Z=27------------')
clef=2
eachcar=input("entrer le caractere en majuscule: ")
ciphred=(ord(eachcar)-65+int(clef))%26
print("voici le chiffrement de la lettre modulo 26:",ciphred)
print("le caractere chiffré est: ",chr(ciphred+65))
```

### 1.4 Parcours d'une chaîne de caractères
> On itère sur chaque lettre de la chaîne et on affiche son code ASCII normalisé.

```python
print("---------parcours d une chaine---------")
chaine = input("donner la chaine?")
for eachcar in chaine:
  print("voici son code ASCII", ord(eachcar))
  print("voici son code ACII normalise sur 26 : ", ord(eachcar)-65)
```

## 2. Chiffrement de César
> **Principe :** Chaque lettre est décalée de **k** positions dans l'alphabet (modulo 26).
> - **Chiffrement :** `C = (P + k) % 26`
> - **Déchiffrement :** `P = (C - k) % 26`
> - Clé : un entier entre 1 et 25

### 2.1 Chiffrement lettre par lettre d'une chaîne (version détaillée)

```python
print("---------chiffrement d une chaine---------")
chaine=input("donnez la chaine a chiffrer?")
clef=input("donnez la clef de chiffrement?")
chainechiffre=""
for eachcar in chaine:
  print("voici le chiffrement de la lettre modulo :",(ord(eachcar)-65+int(clef))%26)
  cipherd=int(((ord(eachcar)-65+int(clef))%26)+65)
  print("voici le caractere chiffré: ",chr(cipherd))
  chainechiffre+=(chr(cipherd))
print("voici le chiffrement des lettres", chainechiffre)
```

### 2.2 Fonction `cesar_chiffrement(chaine, clef)`
> Version encapsulée dans une fonction réutilisable.

```python
def cesar_chiffrement(chaine, clef):
  chainechiffre=""

  for eachcar in chaine:
    cipherd=int(((ord(eachcar)-65+int(clef))%26)+65)
    print("voici le chiffrement de la lettre:", chr(cipherd))
    chainechiffre+=(chr(cipherd))
  return chainechiffre

chaine = input("donnez la chaine a chiffrer?")
clef = input("donnez la clef de chiffrement?")
resultat = cesar_chiffrement(chaine, clef)
print("voici le chiffrement des lettres", resultat)
```

### 2.3 Fonction `cesar_dechiffrement(chaine, clef)`
> Opération inverse : on soustrait la clé au lieu de l'ajouter.

```python
def cesar_dechiffrement(chaine, clef):
  chainedechiffre=""

  for eachcar in chaine:
    dechpherd=int(((ord(eachcar)-65-int(clef))%26)+65)
    print("voici le dechiffrement de la lettre:", chr(dechpherd))
    chainedechiffre+=(chr(dechpherd))
  return chainedechiffre

chaine = input("donnez la chaine a dechiffrer?")
clef = input("donnez la clef de dechiffrement?")
resultat = cesar_dechiffrement(chaine, clef)
print("voici le dechiffrement des lettres", resultat)
```

### 2.4 Attaque par force brute — Test des 26 clés possibles
> César est **vulnérable** : il n'existe que 25 clés différentes.
> On peut tester toutes les clés et trouver le texte original sans connaître la clé.

```python
chaine = input("donnez la chaine a chiffrer?")
for clef in range(1,27):
  resultat=cesar_dechiffrement(chaine, clef)
  print("La clef : ", clef,"donne le texte dechiffre", resultat)
```

## 3. Chiffrement de Vernam (One-Time Pad)
> **Principe :** Chaque lettre est combinée avec la lettre correspondante de la clé (XOR alphabétique).
> - **Chiffrement :** `C[i] = (P[i] + K[i]) % 26`
> - **Déchiffrement :** `P[i] = (C[i] - K[i]) % 26`
> - **Contrainte :** la clé doit avoir **exactement la même longueur** que le texte.
> - **Sécurité :** mathématiquement incassable si la clé est aléatoire et utilisée une seule fois.

### 3.1 Fonction `vernam_chiffrement(text_clair, cle)`

```python
def vernam_chiffrement(text_clair, cle):
  alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  text_chiffre = ""
  for i in range(len(text_clair)):
    lettre_claire = text_clair[i]
    lettre_cle = cle[i]
    index_clair = alphabet.index(lettre_claire)
    index_cle = alphabet.index(lettre_cle)
    index_chiffre = (index_clair + index_cle) % 26
    text_chiffre += alphabet[index_chiffre]
  return text_chiffre
```

### 3.2 Test du chiffrement de Vernam
> Exemple : `AZZOUZI` avec clé `MOHAMMED` → `MNGOGZM`

```python
chaine = input("donnez la chaine a chiffrer?")
clef = input("donnez la clef de chiffrement?")
resultat = vernam_chiffrement(chaine, clef)
print("voici le chiffrement des lettres", resultat)
```

### 3.3 Fonction `vernam_dechiffrement(text_chiffre, cle)`

```python
def vernam_dechiffrement(text_chiffre,cle):
  alphabet ="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  text_dechiffre=""
  for i in range(len(text_chiffre)):
    lettre_chiffree=text_chiffre[i]
    lettre_cle=cle[i]
    index_chiffre = alphabet.index(lettre_chiffree)
    index_cle = alphabet.index(lettre_cle)
    index_dechiffre = (index_chiffre - index_cle) % 26
    text_dechiffre += alphabet[index_dechiffre]
  return text_dechiffre
```

### 3.4 Test du déchiffrement de Vernam
> Vérification : `MNGOGZM` avec clé `MOHAMMED` → `AZZOUZI` ✓

```python
chaine = input("donnez la chaine a dechiffrer?")
clef = input("donnez la clef de dechiffrement?")
resultat = vernam_dechiffrement(chaine, clef)
print("voici le dechiffrement des lettres", resultat)
```

## 4. Interface Menu — Application Complète
> Fonction `menu()` interactive qui regroupe les deux algorithmes.
> L'utilisateur choisit l'algorithme (César ou Vernam) et l'action (chiffrement ou déchiffrement).

### 4.1 Définition de la fonction `menu()`

```python
def menu():
    print("--------MENU CHIFFREMENT-----")
    print("1 - Cesar")
    print("2 - Vernam")

    choix_algo = input("Choisissez l'algorithme : ")

    print("\n1 - Chiffrement")
    print("2 - Dechiffrement")

    choix_action = input("Choisissez l'action : ")

    texte = input("\nEntrez le texte : ")
    cle = input("Entrez la cle : ")

    if choix_algo == "1":
        cle = int(cle)

        if choix_action == "1":
            resultat = cesar_chiffrement(texte, cle)
        elif choix_action == "2":
            resultat = cesar_dechiffrement(texte, cle)
        else:
            print("Choix invalide")
            return

    elif choix_algo == "2":
        if choix_action == "1":
            resultat = vernam_chiffrement(texte, cle)
        elif choix_action == "2":
            resultat = vernam_dechiffrement(texte, cle)
        else:
            print("Choix invalide")
            return

    else:
        print("Algorithme invalide")
        return

    print("\nResultat :", resultat)
```
